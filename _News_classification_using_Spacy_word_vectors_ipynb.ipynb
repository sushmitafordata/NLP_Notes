```python
df['label_num']=df['label'].map({'Fake' : 0, 'Real': 1})
df.head()
```

```python
import spacy
nlp=spacy.load("en_core_web_lg")
```

```python
doc=nlp("Top Trump Surrogate Brutally Stabs Hi")
doc.vector
```

- As you can see using spacy we can create vector for any particular sentence.
- SO we are going to create a new column which will have vectors of different sentences.

```python
df['vector']=df['Text'].apply(lambda x: nlp(x).vector)
df.head()
```

- Every sentence has been converted into 300 element vector .
- Now we are going to use this vector to train our model.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df.vector.values,df.label_num,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test
```

- The data we are getting is array of array.
- So we are going to convert it into 2d numpy array using np.stack

```python
import numpy as np

X_train_2d =np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
X_train_2d
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler


"""the fit() method will allow us to get the parameters of the scaling function.
The transform() method will transform the dataset to proceed with further data analysis steps.
The fit_transform() method will determine the parameters and transform the dataset."""

#Since our data has negative values as well it is going to throw as error with MultinomialNB model.
# To avoid the error we will use MinMax Scaler

scaler=MinMaxScaler()
scaler_train_embed= scaler.fit_transform(X_train_2d)
scaler_test_embed=scaler.transform(X_test_2d)

clf=MultinomialNB()
clf.fit(scaler_train_embed,y_train)
```

```python
from sklearn.metrics import classification_report

y_pred=clf.predict(scaler_test_embed)
print(classification_report(y_test,y_pred))
```

```python
from  sklearn.neighbors import KNeighborsClassifier

#1. creating a KNN model object
clf = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')

#2. fit with all_train_embeddings and y_train
clf.fit(X_train_2d, y_train)

#3. get the predictions for all_test_embeddings and store it in y_pred
y_pred = clf.predict(X_test_2d)

#4. print the classfication report
print(classification_report(y_test, y_pred))
```
